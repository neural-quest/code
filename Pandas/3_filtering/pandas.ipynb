## Tutorial-3: Filtering

```python
df.head()
```

```python
filter_1 = df["winner"] == "Chennai Super Kings"
```

```python
type(filter_1)
```

```python
filter_1
```

```python
df[filter_1]
```

```python
df[df["winner"]=="Chennai Super Kings"]
```

```python
df.loc[filter_1]
```

```python
df.loc[filter_1, ["team1", "team2", "winner"]]
```

```python
& and
| or
```

```python
filter_2 = (df["winner"] == "Chennai Super Kings") & (df["city"] == "Chennai")
```

```python
df.loc[filter_2, ["city", "team1", "team2", "winner"]]
```

```python
filter_3 = (df["team1"] == "Mumbai Indians") | (df["team2"] == "Mumbai Indians")
```

```python
df.loc[filter_3, ["team1", "team2"]]
```

```python
df.loc[~filter_3,  ["team1", "team2"]]
```

```python
filter_4 = (df["city"] =="Mumbai") | (df["city"] =="Delhi") | (df["city"] =="Chennai")

df.loc[filter_4]
```

```python
city = ["Mumbai", "Delhi", "Chennai"]

filter_5 = df["city"].isin(city)
```

```python
df.loc[filter_5]
```
